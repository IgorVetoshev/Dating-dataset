```python
from time import sleep
from time import time
from random import randint
from IPython.display import clear_output

# Redeclaring the lists to store data in
names = []
scores = []

# Preparing the monitoring of the loop
start_time = time()
requests = 0
pages = [str(i) for i in range(1,8)]

# For every page in the interval 1-4
for page in pages:

    # Make a get request
    response = get('https://www.kinopoisk.ru/user/2340770/votes/list/ord/date/page/' + page + '/#list')

    # Pause the loop
    sleep(randint(8,15))

    # Monitor the requests
    requests += 1
    elapsed_time = time() - start_time
    print('Request:{}; Frequency: {} requests/s'.format(requests, requests/elapsed_time))
    clear_output(wait = True)


    # Parse the content of the request with BeautifulSoup
    page_html = BeautifulSoup(response.text, 'html.parser')

    # Select all the 50 movie containers from a single page
    movie_containers = page_html.find_all('div', class_ = 'item even')

    # For every movie of these 50

    for container in movie_containers:

        # The name
        name = container.a.text
        names.append(name)

        # The score
        score = container.find('div', class_ = 'vote').text
        scores.append(score)

import pandas as pd

test_df = pd.DataFrame({'movie': names,
                       'score': scores})
print(test_df.info())
test_df
```
